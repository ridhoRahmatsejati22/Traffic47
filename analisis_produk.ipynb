```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

# Baca dataset
file_path = 'women_clothing_ecommerce_sales.csv'
data = pd.read_csv(file_path)

# Konversi kolom order_date ke format datetime
data['order_date'] = pd.to_datetime(data['order_date'])

# Mengelompokkan data berdasarkan tanggal dan SKU untuk mendapatkan penjualan harian per produk
daily_sales = data.groupby(['order_date', 'sku'])['quantity'].sum().reset_index()

# Daftar SKU yang akan diprediksi
sku_list = daily_sales['sku'].unique()

# Loop untuk memprediksi semua SKU dalam daftar
for sku_to_predict in sku_list:
    sku_sales = daily_sales[daily_sales['sku'] == sku_to_predict]

    # Pastikan ada cukup data untuk melakukan prediksi
    if len(sku_sales) < 2:
        print(f"Tidak ada cukup data untuk SKU {sku_to_predict} untuk melakukan prediksi.")
    else:
        # Siapkan data untuk Prophet
        df_prophet = sku_sales[['order_date', 'quantity']].rename(columns={'order_date': 'ds', 'quantity': 'y'})

        # Inisialisasi model Prophet
        model = Prophet()

        # Fit model
        model.fit(df_prophet)

        # Buat dataframe untuk prediksi (30 hari ke depan)
        future = model.make_future_dataframe(periods=30)

        # Prediksi
        forecast = model.predict(future)

        # Plot hasil prediksi
        fig = model.plot(forecast)
        plt.title(f'Sales Forecast for {sku_to_predict}')
        plt.xlabel('Date')
        plt.ylabel('Sales Quantity')
        plt.show()

        # Plot komponen prediksi
        fig2 = model.plot_components(forecast)
        plt.show()

        # Rekomendasi jumlah stok optimal (misalnya berdasarkan prediksi median)
        recommended_stock = forecast['yhat'].tail(30).sum()
        print(f'Rekomendasi jumlah stok optimal untuk {sku_to_predict} selama 30 hari ke depan: {recommended_stock}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset
file_path = 'women_clothing_ecommerce_sales.csv'
data = pd.read_csv(file_path)

# Konversi kolom order_date ke format datetime
data['order_date'] = pd.to_datetime(data['order_date'])

# Mengelompokkan data berdasarkan SKU untuk mendapatkan total penjualan per produk
product_sales = data.groupby('sku')['quantity'].sum().reset_index()

# Mengurutkan produk berdasarkan total penjualan (dari yang tertinggi)
product_sales = product_sales.sort_values(by='quantity', ascending=False)

# Ambil 10 produk teratas (atau sesuai dengan kebutuhan)
top_products = product_sales.head(10)

# Plot grafik untuk setiap produk
for index, row in top_products.iterrows():
    sku = row['sku']
    product_data = data[data['sku'] == sku]

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(product_data['order_date'], product_data['quantity'], marker='o', linestyle='-')
    plt.title(f'Penjualan Produk {sku} Seiring Waktu')
    plt.xlabel('Tanggal')
    plt.ylabel('Jumlah Terjual')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

# Save Model

```python
import pandas as pd
from prophet import Prophet
import pickle

# Baca dataset
file_path = 'women_clothing_ecommerce_sales.csv'
data = pd.read_csv(file_path)

# Konversi kolom order_date ke format datetime
data['order_date'] = pd.to_datetime(data['order_date'])

# Mengelompokkan data berdasarkan tanggal dan SKU untuk mendapatkan penjualan harian per produk
daily_sales = data.groupby(['order_date', 'sku'])['quantity'].sum().reset_index()

# Daftar SKU yang akan diprediksi
sku_list = daily_sales['sku'].unique()

# Membuat dictionary untuk menyimpan model Prophet untuk setiap SKU
prophet_models = {}

# Loop untuk memprediksi semua SKU dalam daftar
for sku_to_predict in sku_list:
    sku_sales = daily_sales[daily_sales['sku'] == sku_to_predict]

    # Pastikan ada cukup data untuk melakukan prediksi
    if len(sku_sales) < 2:
        print(f"Tidak ada cukup data untuk SKU {sku_to_predict} untuk melakukan prediksi.")
    else:
        # Siapkan data untuk Prophet
        df_prophet = sku_sales[['order_date', 'quantity']].rename(columns={'order_date': 'ds', 'quantity': 'y'})

        # Inisialisasi model Prophet
        model = Prophet()

        # Fit model
        model.fit(df_prophet)

        # Simpan model Prophet ke dalam dictionary
        prophet_models[sku_to_predict] = model

# Simpan model Prophet ke dalam file menggunakan pickle dengan nama file 'analisis_produk.sav'
filename = 'analisis_produk.sav'
with open(filename, 'wb') as file:
    pickle.dump(prophet_models, file)
```
